```python
import tbapy
import numpy as np
import os
import seaborn as sns
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import warnings
warnings.filterwarnings('ignore') 
import pickle
import numpy as np
import pandas as pd
```

```python
tba = tbapy.TBA(os.environ['TBAKEY'])

event = tba.event_matches('2017alhu')
```

```python
def get_comp(team, tba):
    comps = tba.team_events(team, year=2017, keys=True)
    if(len(comps)==0):
        return (0, 0)
    opr = tba.event_oprs(comps[-1])['oprs'][team]
    return (opr, len(comps))
```

```python
arr = {'team':[], 'opr':[], 'comps':[]}
for i in tba.event_teams('2018alhu', keys=True):
    arr['team'].append(i)
    arr['opr'].append(get_comp(i, tba)[0])
    arr['comps'].append(get_comp(i, tba)[1])
```

```python
data = pd.DataFrame(arr)
data = data.set_index('team')
```

```python
p1 = sns.stripplot(y='comps', x='opr', orient="h", jitter=True, data=data)

p1.set_title('Offensive Power Rating')
x = p1.get_ylim()
p1.plot(len(x) * [data['opr']['frc4924']], x, color='red')
plt.show()
```
